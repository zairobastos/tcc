1. Importação das bibliotecas utilizadas

```python
import pandas as pd
```

2. Leitura do CSV

```python
df_supermarket = pd.read_csv('./data/sales_curva_a.csv', delimiter=',', encoding='UTF-8',low_memory=False)
df_supermarket.head()
```

3. listando os produtos do dataset

```python
produtos = df_supermarket['item_nbr'].unique().tolist()
print('Produtos:', produtos)
print('Quantidade de produtos:', len(produtos))
```

4. Separando o date em ano, mes e dia

```python
df_supermarket.loc[:, 'ano'] = pd.to_datetime(df_supermarket['date']).dt.year
df_supermarket.loc[:, 'mes'] = pd.to_datetime(df_supermarket['date']).dt.month
df_supermarket.loc[:, 'dia'] = pd.to_datetime(df_supermarket['date']).dt.day
df_supermarket.loc[:, 'dia_semana'] = pd.to_datetime(df_supermarket['date']).dt.dayofweek
df_supermarket.head()
```

5. Selecionando algumas colunas

```python
df_supermarket = df_supermarket[['item_nbr', 'unit_sales','price', 'dia', 'mes', 'ano', 'day_of_year','date']]
df_supermarket.sample(5)
```

6. Filtrando o produto e o ano

```python
df_supermarket.loc[(df_supermarket['item_nbr'] == 550) & (df_supermarket['ano'] == 2019)]
```
